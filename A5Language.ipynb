```python
import os
from google.colab import userdata

!pip -q install openai

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
from openai import OpenAI
client = OpenAI()
```

```python
prompt = """
you are inventing a fictional language spoken by animals in the jungle.
these are not real words and should not be based on real words.
each phrase must be 2-5 words.
include repeating sylables and soft consonants.
return exactly 16 phrases, one per line, no numbering, no punctuation other than spaces.
"""

response = client.responses.create(
    model="gpt-5-mini",
    input=prompt
)

phrases = response.output_text.strip()
print(phrases)
```

```python
from google.colab import files

with open("animal_language.txt", "w", encoding="utf-8") as f:
    f.write(phrases)

files.download("animal_language.txt")
```
